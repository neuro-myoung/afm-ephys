# Stacked Traces

The following code will allow you to generate stacked traces from individual sweeps. The PlotData class is generated from an augmented data file that can be called from its path without the file suffix ('_augmented.csv') included.
The class has the following defined methods that can be used to process the PlotData object:

    
    plot_sweep(sweep, vars, roi=None, scalebars=False, scalelabs=False)
        Arguments:
            sweep: sweep number to be plotted
            vars: list of traces to be included in the stacked plot (can be position, force, work, and or i_blsub)
            roi: Allows one to define a window of the trace to show as opposed to the entire trace. This argument must be passed as an iterable (list or tuple). Default None will show the entire sweep.
            scalebars: Allows automated generation of scalebars that are 20% the vertical height of the window and 100 ms for y and x respectively. Defaults to False which will not show scalebars.
            scalelabs: Allows the automatic generation of appropriate unit labels for the scalebars. Defaults to False whcih will not show labels.
            
        Output:
            A plot of stacked traces.
            
      remove_sweep(sweeps)
          Arguments:
              sweeps: Can be a list of sweeps or a single sweep number. 
              
          Output: A new PlotData object with the sweeps passed to the function removed.
          
          

```python
import plot_data as plot

dat = plot.PlotData('mp2/20190816_hek293T_mp2_c2')
dat.plot_sweep(4, vars = ['position','force','work','i_blsub'], roi = [300,1100], scalebars=True, scalelabs=True)
```
